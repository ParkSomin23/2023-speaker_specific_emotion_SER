```python
import numpy as np
import pandas as pd
import os
import glob as glob
from tqdm import tqdm
import yaml
```

```python
p_path = '/home/ubuntu/Dropbox/23_for_revision/with_emotion/Vox_IEMO/'
```

```python
hparams_file = glob.glob(p_path + '*only_emo*/*/*/hparams.yaml') + glob.glob(p_path + '*only_emo*/*/hparams.yaml')
hparams_file.sort()
```

```python
_dict = { 
        'folder':[],
        'numbers':[]
        }

for h in hparams_file:
    if h.split('/')[-3] == 'tmp':
        continue    

    with open(h) as f:
        hparams = yaml.load(f, Loader=yaml.FullLoader)

    _dict['folder'].extend([h.split('/')[-3]])
    _dict['numbers'].extend([h.split('/')[-2]])
    for k in hparams.keys():
        if k not in _dict.keys():
            _dict[k] = []
            
        if k == 'folder':
            continue
        if k == 'numbers':
            continue
        
        try:
            _dict[k].extend([hparams[k]])
        except:
            _dict[k].extend(['None'])
```
